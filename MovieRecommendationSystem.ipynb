```python
path1 = "/content/drive/MyDrive/Movie Recommendation System/tmdb_5000_movies.csv"
path2 = "/content/drive/MyDrive/Movie Recommendation System/tmdb_5000_credits.csv"
movies = pd.read_csv(path1)
credits = pd.read_csv(path2)
```

```python
movies = movies.merge(credits , on = 'title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(3)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

We here remove any spaces between names to avoid confusion .
For Eg : There might be an interpretation error between "Sam Worthington" and "Sam Mendes" since the first names are same.

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_Movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_Movies['tags'] = new_Movies['tags'].apply(lambda x: " ".join(x))
new_Movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector_Movies = cv.fit_transform(new_Movies['tags']).toarray()
```

```python
vector_Movies.shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps_Movies = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps_Movies.stem(i))

  return " ".join(y)
```

```python
new_Movies['tags'].apply(stem)
```

```python
new_Movies['tags'] = new_Movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector_Movies)
```

```python
similarity[1]
```

```python
new_Movies[new_Movies['title'] == 'Batman Begins'].index[0]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True , key = lambda x: x[1])
```

```python
def recommend(movie):
    index = new_Movies[new_Movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_Movies.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
